```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/begum4ozcan/corporate-sustainability.dsa210/main/data/raw/epi2024results.csv"
df = pd.read_csv(url)

df.head()
```

```python
df["EPI_change"] = df["EPI.new"] - df["EPI.old"]

df[["country", "EPI.old", "EPI.new", "EPI_change"]].head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(df["EPI.old"], df["EPI.new"], alpha=0.7)
plt.plot([0,100], [0,100], "r--")
plt.xlabel("EPI (old)")
plt.ylabel("EPI (new)")
plt.title("Change in Environmental Performance Index")
plt.show()
```

```python
df["EPI_change"].hist(bins=30)
plt.xlabel("EPI change")
plt.ylabel("Number of countries")
plt.title("Distribution of EPI Changes (Old â†’ New)")
plt.show()
```

```python
df["EPI_change"].describe()
```

Most of the countries' Environmental Performace Index scores increased from the previous version to 2024. However, the magnitude of change varies remarkably, which indicates uneven progress in environmental performance across countries.
