```python
import pandas as pd
import numpy as np
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/begum4ozcan/corporate-sustainability.dsa210/main/data/raw/epi2024results.csv"
df = pd.read_csv(url)

df.head()
```

```python
df = df.replace("NA", np.nan)

for c in df.columns:
    if c not in ["country", "iso", "code"]:
        df[c] = pd.to_numeric(df[c], errors="coerce")
```

```python
df.info()
df.describe().T.head(10)
```

```python
df.to_csv("epi2024results_clean.csv", index=False)
```
