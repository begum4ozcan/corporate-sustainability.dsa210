# DSA 210 - Introduction to Data Science
**Environmental Performance Index (EPI) Analysis**

**Motivation:**

Environmental sutainability is a critical global issue, and measuring environmental performance across countries helps assess progress and identify challenges.
This project analyzes the EPI to explore changes in evironmental performance and to model how different environmental indicators contribute to a country's overall EPI score.


**Data Source:**

I used the 2024 EPI dataset published by the Yale Center for Environmental Law and Policy for this project.
- Observations: 180 countries
- Variables: EPI scores and multiple environmental indicators of these countries
- Data files: raw, processed


**Data Preparation:**

In 01_data_cleaning.ipynb, I performed data cleaning and included:
- Handling missing values ("NA" -> NaN)
- Converting indicator columns to numeric types
- Verifying data consistency and structure


**Exploratory Data Analysis:**

Exploratory analysis was conducted in 02_eda.ipynb.
Key steps:
- Sceated a new variable EPI_change = EPI.new - EPI.old
- visualized old vs. new EPI scores
- examined the distibution of EPI changes
Finding:
Most countries experienced an increase inEPI scores, though the magnitude of improvement varies.

**Machine Learning:**

Machine learning models were implemented in ML_models.ipynb to predict EPI.new.
- *Target:* EPI.new
- *Features:* Environmental indicators ending with .new
- *Models:* Ridge Recession, Random Forest
- *Evaluation:* 5-fold cross validation (RMSE, R^2)

**Result:**
Ridge recession achieved very high predictive performance, which is expected since EPI is a composite index constructed from the pedictor indicators. Random Forest performed slightly worse, suggesting largely linear relationship.


**Feature Importance:**

Permutation importance analysis shows that Ecosystem Vitality(ECO.new) is the strongest contributor to the EPI score, followed by indicators related to health, governance, and pollution. This alligns with the official EPI methodology.


**Conclusion:**

This project demonstrates how data cleaning, exploratory analysis, and machine learning can be applied to environmental sustainability data, providing insights into the structure and interpretation of the EPI.
